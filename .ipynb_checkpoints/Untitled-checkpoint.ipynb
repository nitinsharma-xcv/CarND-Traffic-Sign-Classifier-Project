````python
import pickle
import matplotlib.pyplot as plt
from skimage.transform import rotate

training_file = 'data/train.p'
with open(training_file, mode='rb') as f:
    train = pickle.load(f)
    
X_train, y_train = train['features'], train['labels']

_ = plt.subplots(figsize=(20, 5))
ax = plt.subplot(131)
ax.set_title('original image')
im = plt.imread('web_images/web7_32x32.jpg')
ax.imshow(im)

ax = plt.subplot(132)
ax.set_title('+10 degree rotated image')
im2 = rotate(im,angle=20)
ax.imshow(im2)

ax = plt.subplot(133)

im3 = rotate(im,angle=-10)
ax.imshow(im3)
````
